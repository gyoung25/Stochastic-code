```python
# Random walk
import numpy as np
import matplotlib.pyplot as plt

k=1
maxIt=5000;
xy=np.zeros((maxIt,2))

#pr is the probability of moving right
pr=0.25
pl=0.25
pu=0.25
pd=1-pl-pr-pu

if pd<0:
    print('Probabilities bad')
else:
    while k<maxIt:
        #generate random number between 0 and 1
        q=np.random.uniform(0,1)
        
        #construct probability intervals
        r=pr
        l=r+pl
        u=l+pu
        d=u+pd
        
        #step left, right, up, or down
        xy[k,0]=xy[k-1,0]+(q<r)-(q>=r)*(q<l)
        xy[k,1]=xy[k-1,1]+(q>=l)*(q<u)-(q>=u)*(q<=d)
        k+=1

x=xy[:,0]
y=xy[:,1]
N=1.1*np.max(abs(xy))

plt.plot([-N, N],[0, 0],color='k')
plt.plot([0, 0],[-N, N],color='k')
plt.plot(x,y,linewidth=2)
plt.axis([-N, N, -N, N])
plt.show()
```
